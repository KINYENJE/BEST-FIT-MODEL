```python
import pandas as pd
import numpy as np

from scipy.optimize import curve_fit
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#Load dataset in txt format

date, temperature = np.loadtxt('annual.txt', unpack=True, usecols=(1,2), skiprows=1)
```

```python
plt.plot(date,temperature,'.')
plt.ylabel("Mean Temperatures", fontsize=15)
plt.xlabel("Years", fontsize=15)
```

```python
#Calculate the slope and intercept of the line of best

slope, intercept = np.polyfit(date,temperature,1)
```

```python
#Create a new set of date-values to represent the line of best fit

date_fit = np.linspace(date.min(), date.max(), 100)
```

```python
# Calculate the corresponding temperature-values
#  for the line of best fit using the slope and intercept values:

temp_fit = slope* date_fit + intercept
```

```python
# Plot the original data points
plt.plot(date, temperature, 'o')

# plot the line of best fit
plt.plot(date_fit, temp_fit, '-')

# axis labels
plt.ylabel("Mean Temperatures", fontsize=15)
plt.xlabel("Years", fontsize=15)
plt.show()
```
